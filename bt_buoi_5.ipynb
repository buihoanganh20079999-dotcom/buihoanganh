```python
#1
def Max_of_Int(a,b) -> int:
  return [a if a>b else b]
a, b = map(int, input().split())
print(Max_of_Int(a,b))
```

```python
#2
def swap(a, b):
  return b,a
a,b=map(int, input().split())
print(*swap(a,b))
```

```python
#3
a = int(input())
def snt(a):
  for i in range(2,a):
    if a%i==0:
      return False
    return True
print(snt(a))
```

```python
#4
a = int(input())
def so_hoan_hao(a):
  tong_cac_uoc = 0
  for i in range(1, a):
    if a % i == 0:
      tong_cac_uoc += i
  if tong_cac_uoc == a:
    return ('True')
  else:
    return ('False')
print(so_hoan_hao(a))
```

```python
#5
def vi_tri_so():
  a = list(map(int,input('Nhap day so: ').split()))
  b = int(input('Nhap so: '))
  for i in range(len(a)):
    if a[i] == b:
      return i + 1
    return -1
vi_tri_so()
```

```python
#9
def cal(x):
  res = 0
  while x != 0:
    res += x % 10
    x //= 10
  return res
n = int(input())
print(cal(n))
```

```python
#10
def is_anhxa(a: str, b: str) -> bool:
    map_a_to_b = {}
    map_b_to_a = {}
    for ch1, ch2 in zip(a,b):
        if ch1 in map_a_to_b:
            if map_a_to_b[ch1] != ch2:
                return False
        else:
            map_a_to_b[ch1] = ch2

        if ch2 in map_b_to_a:
            if map_b_to_a[ch2] != ch1:
                return False
        else:
            map_b_to_a[ch2] = ch1

    return True
a, b = str(input("Nhập 2 chữ a và b: ")).split()
print(is_anhxa(a,b))
```
