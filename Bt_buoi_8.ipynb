```python
#1
pi = 3.14
class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def total_area(self):
        return 2*pi*self.radius*self.height + 2*pi*self.radius**2
    def volume(self):
        return pi*self.radius**2*self.height

r = int(input()))
h = int(input())
t = Cylinder(r,h)
print (f"Total Area: {t.total_area():.2f}")
print (f"Volume: {t.volume():.2f}")
```

```python
#2
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
    def is_leap_year(self):
      leap = False
      if self.year % 400 == 0:
            leap = True
      if self.year % 4 == 0 and self.year % 100 != 0:
            leap = True
      return leap
    def day_in_month(self):
        days = 0
        if self.month in [1,3,5,7,8,10,12]:
          days = 31
        if self.month == 2:
          if self.is_leap_year():
            days = 29
          else:
            days = 28
        else:
            days = 30
        return days
    def is_valid(self):
      valid = True
      if self.month < 1 or self.month > 12:
        valid = False
      if valid:
        max_day = self.day_in_month()
        if self.day < 1 or self.day > max_day:
          valid = False
      return valid
    def next_day(self):
      self.day += 1
      max_day = self.day_in_month()

      if self.day > max_day:
        self.day = 1
        self.month +=1

        if self.month > 12:
          self.month = 1
          self.year += 1

s = input().strip()
day, month, year = map(int, s.split("/"))
date1 = Date(day, month, year)
if not date1.is_valid():
  print ("Invalid")
else:
  date1.next_day()
  print (f"{date1.day:02d}/{date1.month:02d}/{date1.year}")
```

```python
#3
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def check(self):
    if self.x != 0 and self.y != 0:
      return "Toạ độ điểm M không thuộc gốc toạ độ"
    elif self.x > 0 and self.y == 0:
      return "Toạ độ điểm M thuộc trục hoành"
    elif self.y > 0 and self.x == 0:
      return "Toạ độ điểm M thuộc trục tung"
    else:
      return "Điểm M thuộc gốc toạ độ"

  def check_distance(self):
    result_square = self.x ** 2 + self.y ** 2
    result = result_square ** 0.5
    return round(result, 2)

x = float(input("Nhập toạ độ x: "))
y = float(input("Nhập toạ độ y: "))
point1 = Point(x,y)
print(point1.check())
print(point1.check_distance())
```

```python
#5
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        self.items.append({"name": name, "price": price})

    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                return
        print("Không tìm thấy sản phẩm")

    def is_empty(self):
        return len(self.items) == 0

    def total_price(self):
        total = 0
        for item in self.items:
            total += item["price"]
        return total

    def show_items(self):
        if self.is_empty():
            print("Giỏ hàng rỗng")
        else:
            for item in self.items:
                print(f"{item['name']} - {item['price']}")

    def clear(self):
        self.items.clear()


cart = ShoppingCart()

while True:
    print("\n--- MENU ---")
    print("2. Thêm sản phẩm")
    print("3. Xóa sản phẩm theo tên")
    print("4. Kiểm tra giỏ hàng rỗng")
    print("5. Tính tổng tiền")
    print("6. Hiển thị giỏ hàng")
    print("7. Xóa toàn bộ giỏ hàng")

    choice = input("Chọn chức năng (2-7): ")

    if choice == "2":
        name = input("Tên sản phẩm: ")
        price = float(input("Giá: "))
        cart.add_item(name, price)

    elif choice == "3":
        name = input("Tên sản phẩm cần xóa: ")
        cart.remove_item(name)

    elif choice == "4":
        print("Giỏ hàng rỗng" if cart.is_empty() else "Giỏ hàng không rỗng")

    elif choice == "5":
        print(f"Tổng tiền: {cart.total_price():.2f}")

    elif choice == "6":
        cart.show_items()

    elif choice == "7":
        cart.clear()
        print("Đã xóa toàn bộ giỏ hàng")

    else:
        print("Lựa chọn không hợp lệ")

    exit_choice = input("Có muốn thoát khỏi chương trình không? ")
    if exit_choice.lower() == "yes":
        break
```

```python
#6
class Rectangle:
  def __init__ (self, w, h, k):
    self.w = w
    self.h = h
    self.k = k
  def scale(self):
    return Rectangle(self.w * self.k, self.h * self.k, self.k)
  def area(self):
    return self.w * self.h
  def perimeter(self):
    return (self.w + self.h) * 2

w, h, k = map(float, input().split())
rectangle = Rectangle(w,h,k)
res = rectangle.scale() # res is now a new Rectangle object (scaled)
print (res.area())
print (res.perimeter())
```

```python
#7
from math import gcd
class Fraction:
  def __init__ (self, num, den):
    if den < 0:
      num = -num
      den = -den

    g = gcd (num, den)
    self.num = num // g
    self.den = den // g

  def add (self, other):
    num = self.num * other.den + self.den * other.num
    den = self.den * self.other
    return Fraction (num, den)
  def sub (self, other):
    num = self.num * other.den - self.den * other.num
    den = self.den * self.other
    return Fraction (num, den)
  def mul (self, other):
    num = self.num * other.num
    den = self.den * other.den
    return Fraction (num, den)
  def divide(self, other):
    num = self.num * other.den
    den = self.den * other.num
    return Fraction (num, den)
  def __str__(self):
    return f"{self.num}/{self.den}"

a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)
f1 = Fraction (a, b)
f2 = Fraction (c, d)

if op == "+":
  result = f1.add(f2)
elif op == "-":
  result = f1.sub(f2)
elif op == "*":
  result = f1.mul(f2)
elif op == "/":
  result = f1.divide(f2)

print(result)
```

```python
#8
class BankAccount:
  def __init__ (self, owner, balance):
    self.owner = owner
    self.balance = balance

  def deposit(self, x):
    self.balance += x
  def withdraw(self, x):
    if self.balance >= x:
      self.balance -= x
  def __str__ (self):
    return f"{self.owner},{self.balance}"

owner, balance = input().split()
balance = float(balance)
acc = BankAccount(owner, balance)
n = int(input())
for _ in range(n):
  cmd, x = input().split()
  x = float(x)
  if cmd == "deposit":
    acc.deposit(x)
  if cmd == "withdraw":
    acc.withdraw(x)

print(f"{acc.balance:.2f}")
```

```python
#9
class Student:
  def __init__ (self, name):
    self.name = name
    self.scores = []
  def add_score (self, subject, score):
    self.scores.append(score)

  def avg(self):
    return sum(self.scores) / len(self.scores)
  def rank(self):
    avg = self.avg()
    if avg >= 8.0:
      return "Excellent"
    elif avg >= 6.5:
      return "Good"
    elif avg >= 5:
      return "Average"
    else:
      return "Poor"



name = input().strip()
n = int(input())
stu = Student(name)
for _ in range (n):
  subject, score = input().split()
  score = float(score)
  stu.add_score(subject, score)
print(f"{stu.name} {stu.avg():.2f} {stu.rank()}")
```

```python
#10
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        cnt = 0
        for b in self.books:
            if b.author == name:
                cnt += 1
        return cnt

    def find_by_year(self, y):
        cnt = 0
        for b in self.books:
            if b.year == y:
                cnt += 1
        return cnt


n = int(input())
lib = Library()

for _ in range(n):
    line = input()

    if line.startswith("ADD"):
        _, data = line.split(" ", 1)
        title, author, year = data.split(";")
        year = int(year)

        book = Book(title, author, year)
        lib.add(book)

    elif line.startswith("COUNTYEAR"):
        _, year = line.split()
        print(lib.find_by_year(int(year)))

    elif line.startswith("COUNT"):
        _, author = line.split(" ", 1)
        print(lib.count_by_author(author))
```
